```python

"""ESTABLISH FACTS FROM LABELED DATA: CREATING FEEDBACK MATRIX- without property noun"""

import pandas as pd
import numpy as np
import ast
import os

# Load the dataset
# file_path = 'your_dataset.csv'  # Replace with your actual file path
# data = pd.read_csv(file_path, delimiter='\t')  # Assuming tab-separated file
action_df = pd.read_csv("/home/nguyennguyen/Documents/egotopo/epic-kitchens-55-annotations/EPIC_train_action_labels.csv")
data = action_df

# Define logical predicates as columns
logical_predicates = [
    "UsedFor",
    "CapableOf",
    "CreatedBy",
    "Affordance",
    "HasA",
    "IsA",
    "AtLocation",
    "LocatedNear",
    
]

# Create a set of unique object-object and object-action pairs as rows
data["all_nouns"] = data["all_nouns"].apply(ast.literal_eval)
object_pairs = set()

for index, row in data.iterrows():
    # Create object-object pairs
    if len(row['all_nouns']) > 1:
        for noun1 in row['all_nouns']:
            for noun2 in row['all_nouns']:
                if noun1 != noun2:
                    object_pairs.add((noun1.strip(), noun2.strip()))
    else:
        # Create object-action pairs
        object_pairs.add((row['noun'].strip(), row['verb'].strip()))
print(object_pairs)


# Convert to a list for indexing
object_pairs = list(object_pairs)

# Create a matrix with object pairs as rows and predicates as columns
matrix = pd.DataFrame(0, index=pd.MultiIndex.from_tuples(object_pairs, names=["Object/Action_1", "Object/Action_2"]),
                      columns=logical_predicates)


# Function to populate the matrix based on predefined logic
def populate_matrix(matrix, data):
    for index, row in data.iterrows():
        # Example logic to fill the matrix based on data conditions
        nouns = row['all_nouns']
        # nouns = row['all_nouns'].strip()
        verb = row['verb'].strip()

        # Check logical conditions and populate the matrix
        # Multiple nouns
        if len(nouns) > 1:
            for noun1 in nouns:
                noun1 = noun1.strip()
                for noun2 in nouns:
                    noun2 = noun2.strip()
                    if noun1 != noun2:
                        if (noun1, noun2) in matrix.index:
                            # Example condition: if 'knife' and 'cutting' appear in the same row, mark it in the matrix
                            # if 'knife' in [noun1, noun2] and 'cutting' in [noun1, noun2]:
                            # matrix.at[(noun1, noun2), "AssociatedWith"] = 1

                            # Example condition for hasProperty
                            # if 'lid' in [noun1, noun2] and 'pot' in [noun1, noun2]:
                            matrix.at[(noun1, noun2), "LocatedNear"] = 1
            # Check object-action pairs
            if (noun1, verb) in matrix.index:
                matrix.at[(noun1, verb), "UsedFor"] = 1
                matrix.at[(noun1, verb), "Affordance"] = 1

                matrix.at[(noun2, verb), "UsedFor"] = 1
                matrix.at[(noun2, verb), "Affordance"] = 1
            #else create
        #Single Noun
        else:
            matrix.at[(nouns[0], verb), "UsedFor"] = 1
            matrix.at[(nouns[0], verb), "Affordance"] = 1

# Populate the matrix with logical conditions based on your dataset
populate_matrix(matrix, data)
matrix = matrix.fillna(0)
# Display or save the resulting matrix
path = "/home/nguyennguyen/Documents/egotopo/ego-topo/object_detection/yolov11/predicate_matrix"
file = 'output_matrix_no_prop.csv'
print(matrix)
out_path = os.path.join(path, file)
matrix.to_csv(out_path)
```

```python
"""MATRIX FACTORIZATION: Non-negative Matrix Factorization
 trained using the initial data, and they capture the underlying structure of the relationships between objects/actions and logical predicates.

"""


from sklearn.decomposition import NMF

# Apply NMF
"""
---Adjust n_components to change the dimensionality of the embeddings.
---Add regularization terms if needed by using more advanced libraries like PyTorch or TensorFlow for custom training loops with matrix factorization.
"""
nmf = NMF(n_components=2, init='random', random_state=42)  # `n_components` sets the embedding size
W = nmf.fit_transform(matrix)  # W contains row embeddings
H = nmf.components_.T          # H contains column embeddings

# Display results
print("Row Embeddings (W) - object/action pairs:")
print(W)
print("Column Embeddings (H) - logical predicates:")
print(H)
```
